```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset('iris')
avg_sepal_length = iris.groupby('species')['sepal_length'].mean()
avg_sepal_length.plot(kind='bar', color='skyblue')
plt.title('Average Sepal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.xticks(rotation=45)
plt.show()
#__________________PIE CHART___________________
species_count = iris['species'].value_counts()
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Species in the Iris Dataset')
plt.show()
#_________________BOX PLOT______________________
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Box Plot of Sepal Length by Species')
plt.show()
#_______________HISTOGRAM______________________
plt.hist(iris['sepal_length'], bins=20, color='green', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
#__________________LINE CHARTS & SUB PLOTS___________________
x = iris.index
y1 = iris['sepal_length']
y2 = iris['petal_length']
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
# Line Chart 1: Sepal Length
axs[0].plot(x, y1, label='Sepal Length', color='blue')
axs[0].set_title('Line Chart: Sepal Length')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Sepal Length (cm)')
# Line Chart 2: Petal Length
axs[1].plot(x, y2, label='Petal Length', color='red')
axs[1].set_title('Line Chart: Petal Length')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Petal Length (cm)')
plt.tight_layout()
plt.show()
#_________________SCATTER PLOTS___________________
plt.scatter(iris['sepal_length'], iris['petal_length'], color='purple')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
